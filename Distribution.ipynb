```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)  # (mean, std, size/shape)
s
```

```python
s.shape
```

```python
a = s.reshape(100,10)
```

```python
s
```

```python
a
```

```python
abs(mu - np.mean(s)) < 0.01
```

```python
abs(sigma - np.std(s, ddof=1)) < 0.01
```

```python
import matplotlib.pyplot as plt
# for inline plots in jupyter
%matplotlib inline
# for latex equations
from IPython.display import Math, Latex
# for displaying images
from IPython.core.display import Image
```

```python

count, bins, ignored = plt.hist(s, 30, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')

plt.show()
```

```python
count, bins, ignored = plt.hist(a, 30, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')

plt.show()
```

## Binomial Distribution

```python
n, p = 10, .5  # number of trials, probability of each trial
c = np.random.binomial(n, p, 1000)  # (#trials, probability, size/shape)
# result of flipping a coin 10 times, tested 1000 times.
c
```

```python
c.shape
```

```python
import scipy, scipy.stats
pmf = scipy.stats.binom.pmf(c,10,0.5)
import pylab
pylab.plot(c,pmf)
```

A real world example. A company drills 9 wild-cat oil exploration wells, each with an estimated probability of success of 0.1. All nine wells fail. What is the probability of that happening?

Let’s do 20,000 trials of the model, and count the number that generate zero positive results.

```python
sum(np.random.binomial(9, 0.1, 20000) == 0)/20000
```

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.5,size=1000)
```

```python
ax = sns.distplot(data_binom,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
```
